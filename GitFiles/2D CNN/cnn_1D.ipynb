```python
import os
import librosa
import warnings
import numpy as np
import tensorflow as tf
import pywt  # PyWavelets package for wavelet transforms
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Feature extraction functions
def extract_mfcc(y, sr, n_mfcc=13):
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfcc, axis=1)

def extract_pitch_swipe(y, sr):
    pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
    avg_pitch = np.mean(pitches[magnitudes > np.median(magnitudes)])
    return avg_pitch if not np.isnan(avg_pitch) else 0

def extract_pitch_class_distribution(y, sr):
    pcd = librosa.feature.tonnetz(y=y, sr=sr)
    return np.mean(pcd, axis=1)

def extract_chroma(y, sr):
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    return np.mean(chroma, axis=1)

def extract_dwt(y):
    # Apply discrete wavelet transform
    coeffs = pywt.wavedec(y, 'db1', level=3)
    # Get mean of each coefficient level
    dwt_features = np.hstack([np.mean(c) for c in coeffs])
    return dwt_features

def extract_cwt(y):
    # Apply continuous wavelet transform with different scales
    scales = np.arange(1, 21)
    coefficients, _ = pywt.cwt(y, scales, 'morl')
    # Return mean across time for each scale
    return np.mean(coefficients, axis=1)

# Combine selected features
def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=None)
    features = []

    features.extend(extract_mfcc(y, sr))                            # MFCC (13,)
    features.append(extract_pitch_swipe(y, sr))                    # Pitch (1,)
    features.extend(extract_pitch_class_distribution(y, sr))       # PCD (6,)
    features.extend(extract_chroma(y, sr))                         # Chroma (12,)
    features.extend(extract_dwt(y))                                # DWT (~4, depends on level)
    features.extend(extract_cwt(y))                                # CWT (20, one per scale)

    return np.array(features)

# Suppress warnings
warnings.filterwarnings("ignore")

# Load training dataset (from raga folders)
def load_training_data(data_dir):
    X, y = [], []
    for raga in os.listdir(data_dir):
        raga_path = os.path.join(data_dir, raga)
        if not os.path.isdir(raga_path) or raga.lower() == "test":
            continue
        
        print(f"Processing training data for raga: {raga}")
        file_count = 0
        
        for file in os.listdir(raga_path):
            if file.endswith((".wav", ".mp3")) and ".DS_Store" not in file:
                file_path = os.path.join(raga_path, file)
                try:
                    features = extract_features(file_path)
                    if features is not None:
                        X.append(features)
                        y.append(raga)
                        file_count += 1
                except Exception as e:
                    print(f"⚠️ Error processing {file}: {e}")
        
        print(f"  - Processed {file_count} files for {raga}")
    
    return np.array(X), np.array(y)

# Load test dataset (from Test folder)
def load_test_data(data_dir, encoder):
    test_dir = os.path.join(data_dir, "Test")
    if not os.path.exists(test_dir):
        print(f"⚠️ Test directory not found at {test_dir}")
        return None, None, None
    
    X_test, y_test, file_names = [], [], []
    
    for test_folder in os.listdir(test_dir):
        test_folder_path = os.path.join(test_dir, test_folder)
        if not os.path.isdir(test_folder_path):
            continue
        
        # Extract raga name (remove " Test" suffix if present)
        raga_name = test_folder.replace(" Test", "")
        print(f"Processing test data for raga: {raga_name}")
        file_count = 0
        
        for file in os.listdir(test_folder_path):
            if file.endswith((".wav", ".mp3")) and ".DS_Store" not in file:
                file_path = os.path.join(test_folder_path, file)
                try:
                    features = extract_features(file_path)
                    if features is not None and raga_name in encoder.classes_:
                        X_test.append(features)
                        y_test.append(raga_name)
                        file_names.append(file)
                        file_count += 1
                except Exception as e:
                    print(f"⚠️ Error processing test file {file}: {e}")
        
        print(f"  - Processed {file_count} test files for {raga_name}")
    
    return np.array(X_test), np.array(y_test), file_names

# Set the data directory
data_dir = "../Hybrid/dataset/"

# Load and process training data
X_train, y_train = load_training_data(data_dir)
print(f"✅ Training dataset loaded: {len(X_train)} samples")

# Encode class labels
encoder = LabelEncoder()
y_train_encoded = encoder.fit_transform(y_train)
print(f"🧠 Classes: {encoder.classes_}")

# Normalize training features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# Reshape for CNN input
X_train_reshaped = X_train_scaled.reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)
print(f"📊 Training data shape: {X_train_reshaped.shape}")

# Load and process test data
X_test, y_test, test_files = load_test_data(data_dir, encoder)

if X_test is not None and len(X_test) > 0:
    print(f"✅ Test dataset loaded: {len(X_test)} samples")
    
    # Encode test labels
    y_test_encoded = encoder.transform(y_test)
    
    # Normalize test features using the same scaler
    X_test_scaled = scaler.transform(X_test)
    
    # Reshape for CNN input
    X_test_reshaped = X_test_scaled.reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)
    print(f"📊 Test data shape: {X_test_reshaped.shape}")
else:
    print("⚠️ No test data found or could be processed")
    # Create a small validation split from training data if no test data
    from sklearn.model_selection import train_test_split
    X_train_reshaped, X_test_reshaped, y_train_encoded, y_test_encoded = train_test_split(
        X_train_reshaped, y_train_encoded, test_size=0.2, random_state=42, stratify=y_train_encoded)
    test_files = None

# Build CNN model
model = Sequential([
    Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train_reshaped.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Conv1D(128, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(encoder.classes_), activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary
model.summary()

# Create callbacks for better training
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001)
]

# Train model
history = model.fit(
    X_train_reshaped, 
    y_train_encoded,
    epochs=50,
    batch_size=16,
    validation_data=(X_test_reshaped, y_test_encoded),
    callbacks=callbacks,
    verbose=1
)

# Evaluate model
loss, accuracy = model.evaluate(X_test_reshaped, y_test_encoded)
print(f"🎯 Test Accuracy: {accuracy * 100:.2f}%")

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('raga_classification_history.png')
plt.show()

# Get predictions
y_pred = model.predict(X_test_reshaped)
y_pred_classes = np.argmax(y_pred, axis=1)

# Create confusion matrix
cm = confusion_matrix(y_test_encoded, y_pred_classes)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
           xticklabels=encoder.classes_,
           yticklabels=encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.savefig('raga_confusion_matrix.png')
plt.show()

# Print classification report
print("\nClassification Report:")
class_report = classification_report(
    y_test_encoded, y_pred_classes, 
    target_names=encoder.classes_,
    output_dict=True
)

# Convert to DataFrame for better visualization
df_report = pd.DataFrame(class_report).transpose()
print(df_report)

# If test files are available, save predictions to CSV
if test_files is not None:
    results_df = pd.DataFrame({
        'File': test_files,
        'True Raga': [encoder.classes_[i] for i in y_test_encoded],
        'Predicted Raga': [encoder.classes_[i] for i in y_pred_classes],
        'Correct': y_test_encoded == y_pred_classes
    })
    results_df.to_csv('raga_test_predictions.csv', index=False)
    print("Prediction results saved to 'raga_test_predictions.csv'")

# Save the model
model.save('raga_classification_1d_cnn.h5')
print("Model saved as 'raga_classification_1d_cnn.h5'")
```

```python
import os
import librosa
import warnings
import numpy as np
import tensorflow as tf
import pywt  # PyWavelets package for wavelet transforms
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix, f1_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Suppress warnings
warnings.filterwarnings("ignore")

# Feature extraction functions
def extract_mfcc(y, sr, n_mfcc=13):
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc)
    return np.mean(mfcc, axis=1)

def extract_pitch_swipe(y, sr):
    pitches, magnitudes = librosa.piptrack(y=y, sr=sr)
    avg_pitch = np.mean(pitches[magnitudes > np.median(magnitudes)])
    return avg_pitch if not np.isnan(avg_pitch) else 0

def extract_pitch_class_distribution(y, sr):
    pcd = librosa.feature.tonnetz(y=y, sr=sr)
    return np.mean(pcd, axis=1)

def extract_chroma(y, sr):
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    return np.mean(chroma, axis=1)

def extract_dwt(y):
    # Apply discrete wavelet transform
    coeffs = pywt.wavedec(y, 'db1', level=3)
    # Get mean of each coefficient level
    dwt_features = np.hstack([np.mean(c) for c in coeffs])
    return dwt_features

def extract_cwt(y):
    # Apply continuous wavelet transform with different scales
    scales = np.arange(1, 21)
    coefficients, _ = pywt.cwt(y, scales, 'morl')
    # Return mean across time for each scale
    return np.mean(coefficients, axis=1)

# Combine selected features
def extract_features(file_path):
    y, sr = librosa.load(file_path, sr=None)
    features = []

    features.extend(extract_mfcc(y, sr))                            # MFCC (13,)
    features.append(extract_pitch_swipe(y, sr))                     # Pitch (1,)
    features.extend(extract_pitch_class_distribution(y, sr))        # PCD (6,)
    features.extend(extract_chroma(y, sr))                          # Chroma (12,)
    features.extend(extract_dwt(y))                                 # DWT (~4, depends on level)
    features.extend(extract_cwt(y))                                 # CWT (20, one per scale)

    return np.array(features)

# Get file paths and labels from dataset directory
def get_file_paths_and_labels(data_dir):
    file_paths = []
    labels = []
    
    for raga in os.listdir(data_dir):
        raga_path = os.path.join(data_dir, raga)
        if not os.path.isdir(raga_path):
            continue
        
        print(f"Finding files for raga: {raga}")
        file_count = 0
        
        for file in os.listdir(raga_path):
            if file.endswith((".wav", ".mp3")) and ".DS_Store" not in file:
                file_path = os.path.join(raga_path, file)
                file_paths.append(file_path)
                labels.append(raga)
                file_count += 1
        
        print(f"  - Found {file_count} files for {raga}")
    
    return file_paths, labels

# Load test dataset from Test folder
def load_test_data(test_dir, encoder):
    if not os.path.exists(test_dir):
        print(f"⚠️ Test directory not found at {test_dir}")
        return None, None, None
    
    X_test, y_test, file_names = [], [], []
    
    for test_folder in os.listdir(test_dir):
        test_folder_path = os.path.join(test_dir, test_folder)
        if not os.path.isdir(test_folder_path):
            continue
        
        # Extract raga name (remove " Test" suffix if present)
        raga_name = test_folder.replace(" Test", "")
        print(f"Processing test data for raga: {raga_name}")
        file_count = 0
        
        for file in os.listdir(test_folder_path):
            if file.endswith((".wav", ".mp3")) and ".DS_Store" not in file:
                file_path = os.path.join(test_folder_path, file)
                try:
                    features = extract_features(file_path)
                    if features is not None and raga_name in encoder.classes_:
                        X_test.append(features)
                        y_test.append(raga_name)
                        file_names.append(file)
                        file_count += 1
                except Exception as e:
                    print(f"⚠️ Error processing test file {file}: {e}")
        
        print(f"  - Processed {file_count} test files for {raga_name}")
    
    return np.array(X_test), np.array(y_test), file_names

# Set the data directories
train_data_dir = "../dataset/"
test_data_dir = "../Test/"

# Get file paths and labels for training data
file_paths, labels = get_file_paths_and_labels(train_data_dir)
print(f"✅ Total files found: {len(file_paths)}")

# Split files into train and validation sets first (75% train, 25% validation)
train_paths, val_paths, train_labels, val_labels = train_test_split(
    file_paths, labels, test_size=0.25, random_state=42, stratify=labels
)
print(f"✅ Training set: {len(train_paths)} files")
print(f"✅ Validation set: {len(val_paths)} files")

# Encode class labels
encoder = LabelEncoder()
encoder.fit(labels)  # Fit on all labels to ensure consistent encoding
train_labels_encoded = encoder.transform(train_labels)
val_labels_encoded = encoder.transform(val_labels)
print(f"🧠 Classes: {encoder.classes_}")

# Extract features for training set
print("Extracting features for training set...")
X_train = []
for file_path in train_paths:
    try:
        features = extract_features(file_path)
        if features is not None:
            X_train.append(features)
    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
X_train = np.array(X_train)
print(f"✅ Extracted features for {len(X_train)} training files")

# Extract features for validation set
print("Extracting features for validation set...")
X_val = []
for file_path in val_paths:
    try:
        features = extract_features(file_path)
        if features is not None:
            X_val.append(features)
    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
X_val = np.array(X_val)
print(f"✅ Extracted features for {len(X_val)} validation files")

# Normalize features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)

# Reshape for CNN input
X_train_reshaped = X_train_scaled.reshape(X_train_scaled.shape[0], X_train_scaled.shape[1], 1)
X_val_reshaped = X_val_scaled.reshape(X_val_scaled.shape[0], X_val_scaled.shape[1], 1)
print(f"📊 Training data shape: {X_train_reshaped.shape}")
print(f"📊 Validation data shape: {X_val_reshaped.shape}")

# Load and process test data
X_test, y_test, test_files = load_test_data(test_data_dir, encoder)

if X_test is not None and len(X_test) > 0:
    print(f"✅ Test dataset loaded: {len(X_test)} samples")
    
    # Encode test labels
    y_test_encoded = encoder.transform(y_test)
    
    # Normalize test features using the same scaler
    X_test_scaled = scaler.transform(X_test)
    
    # Reshape for CNN input
    X_test_reshaped = X_test_scaled.reshape(X_test_scaled.shape[0], X_test_scaled.shape[1], 1)
    print(f"📊 Test data shape: {X_test_reshaped.shape}")
else:
    print("⚠️ No test data found or could be processed")
    X_test_reshaped = None
    y_test_encoded = None

# Build CNN model
model = Sequential([
    Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train_reshaped.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Conv1D(128, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(encoder.classes_), activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary
model.summary()

# Create callbacks for better training
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001)
]

# Train model
history = model.fit(
    X_train_reshaped, 
    train_labels_encoded,
    epochs=50,
    batch_size=16,
    validation_data=(X_val_reshaped, val_labels_encoded),
    callbacks=callbacks,
    verbose=1
)

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('raga_classification_history.png')
plt.show()

# Evaluate on test data if available
if X_test_reshaped is not None and y_test_encoded is not None:
    # Get predictions
    y_pred = model.predict(X_test_reshaped)
    y_pred_classes = np.argmax(y_pred, axis=1)
    
    # Calculate metrics
    accuracy = np.sum(y_pred_classes == y_test_encoded) / len(y_test_encoded)
    precision = precision_score(y_test_encoded, y_pred_classes, average='weighted')
    recall = recall_score(y_test_encoded, y_pred_classes, average='weighted')
    f1 = f1_score(y_test_encoded, y_pred_classes, average='weighted')
    
    print(f"🎯 Test Accuracy: {accuracy * 100:.2f}%")
    print(f"🎯 Test Precision: {precision * 100:.2f}%")
    print(f"🎯 Test Recall: {recall * 100:.2f}%")
    print(f"🎯 Test F1 Score: {f1 * 100:.2f}%")
    
    # Create confusion matrix
    cm = confusion_matrix(y_test_encoded, y_pred_classes)
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
               xticklabels=encoder.classes_,
               yticklabels=encoder.classes_)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.savefig('raga_confusion_matrix.png')
    plt.show()
    
    # Print classification report
    print("\nClassification Report:")
    class_report = classification_report(
        y_test_encoded, y_pred_classes, 
        target_names=encoder.classes_,
        output_dict=True
    )
    
    # Convert to DataFrame for better visualization
    df_report = pd.DataFrame(class_report).transpose()
    print(df_report)
    
    # If test files are available, save predictions to CSV
    if test_files is not None:
        results_df = pd.DataFrame({
            'File': test_files,
            'True Raga': [encoder.classes_[i] for i in y_test_encoded],
            'Predicted Raga': [encoder.classes_[i] for i in y_pred_classes],
            'Correct': y_test_encoded == y_pred_classes
        })
        results_df.to_csv('raga_test_predictions.csv', index=False)
        print("Prediction results saved to 'raga_test_predictions.csv'")

# Save the model
model.save('raga_classification_1d_cnn_new.h5')
print("Model saved as 'raga_classification_1d_cnn_new.h5'")
```
