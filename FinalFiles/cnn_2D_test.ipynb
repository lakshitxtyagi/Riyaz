```python
import os
import librosa
import warnings
import numpy as np
import tensorflow as tf
import pywt  # PyWavelets package for wavelet transforms
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix, f1_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Suppress warnings
warnings.filterwarnings("ignore")

# Feature extraction functions for 2D CNN
def extract_melspectrogram(y, sr, n_mels=128, n_fft=2048, hop_length=512):
    """Extract mel spectrogram features suitable for 2D CNN"""
    mel_spec = librosa.feature.melspectrogram(
        y=y, sr=sr, n_mels=n_mels, n_fft=n_fft, hop_length=hop_length
    )
    # Convert to dB scale
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    return mel_spec_db

def extract_mfcc_2d(y, sr, n_mfcc=20, n_fft=2048, hop_length=512):
    """Extract MFCC features suitable for 2D CNN"""
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, n_fft=n_fft, hop_length=hop_length)
    # Add delta and delta-delta features
    mfcc_delta = librosa.feature.delta(mfcc)
    mfcc_delta2 = librosa.feature.delta(mfcc, order=2)
    # Stack together to form a 3-channel image
    return np.stack([mfcc, mfcc_delta, mfcc_delta2], axis=-1)

def extract_chroma_2d(y, sr, n_chroma=12, n_fft=2048, hop_length=512):
    """Extract chromagram features suitable for 2D CNN"""
    chroma = librosa.feature.chroma_stft(y=y, sr=sr, n_chroma=n_chroma, n_fft=n_fft, hop_length=hop_length)
    return chroma

# Combine features for 2D CNN
def extract_features(file_path, fixed_length=256):
    """Extract 2D features for CNN processing with fixed time dimension"""
    y, sr = librosa.load(file_path, sr=22050)  # Fixed sample rate for consistency
    
    # Use a fixed duration by either padding or trimming
    target_length = sr * 10  # Target 5 seconds of audio
    if len(y) < target_length:
        # Pad if shorter
        y = np.pad(y, (0, target_length - len(y)), 'constant')
    else:
        # Trim if longer - take a segment from the middle
        start = len(y) // 2 - target_length // 2
        y = y[start:start + target_length]
    
    # Extract mel spectrogram (main feature for 2D CNN)
    mel_spec = extract_melspectrogram(y, sr, n_mels=128)
    
    # Ensure consistent time dimension
    if mel_spec.shape[1] > fixed_length:
        mel_spec = mel_spec[:, :fixed_length]
    elif mel_spec.shape[1] < fixed_length:
        # Pad with zeros if shorter
        padding = fixed_length - mel_spec.shape[1]
        mel_spec = np.pad(mel_spec, ((0, 0), (0, padding)), 'constant')
    
    # Add channel dimension for CNN
    mel_spec = mel_spec.reshape(mel_spec.shape[0], mel_spec.shape[1], 1)
    
    return mel_spec

# Get file paths and labels from dataset directory
def get_file_paths_and_labels(data_dir):
    file_paths = []
    labels = []
    
    for raga in os.listdir(data_dir):
        raga_path = os.path.join(data_dir, raga)
        if not os.path.isdir(raga_path):
            continue
        
        print(f"Finding files for raga: {raga}")
        file_count = 0
        
        for file in os.listdir(raga_path):
            if file.endswith((".wav", ".mp3")) and ".DS_Store" not in file:
                file_path = os.path.join(raga_path, file)
                file_paths.append(file_path)
                labels.append(raga)
                file_count += 1
        
        print(f"  - Found {file_count} files for {raga}")
    
    return file_paths, labels

# Load test dataset from Test folder
def load_test_data(test_dir, encoder):
    if not os.path.exists(test_dir):
        print(f"⚠️ Test directory not found at {test_dir}")
        return None, None, None
    
    X_test, y_test, file_names = [], [], []
    
    for test_folder in os.listdir(test_dir):
        test_folder_path = os.path.join(test_dir, test_folder)
        if not os.path.isdir(test_folder_path):
            continue
        
        # Extract raga name (remove " Test" suffix if present)
        raga_name = test_folder.replace(" Test", "")
        print(f"Processing test data for raga: {raga_name}")
        file_count = 0
        
        for file in os.listdir(test_folder_path):
            if file.endswith((".wav", ".mp3")) and ".DS_Store" not in file:
                file_path = os.path.join(test_folder_path, file)
                try:
                    features = extract_features(file_path)
                    if features is not None and raga_name in encoder.classes_:
                        X_test.append(features)
                        y_test.append(raga_name)
                        file_names.append(file)
                        file_count += 1
                except Exception as e:
                    print(f"⚠️ Error processing test file {file}: {e}")
        
        print(f"  - Processed {file_count} test files for {raga_name}")
    
    return np.array(X_test), np.array(y_test), file_names

# Set the data directories
train_data_dir = "../dataset/"
test_data_dir = "../Test/"

# Get file paths and labels for training data
file_paths, labels = get_file_paths_and_labels(train_data_dir)
print(f"✅ Total files found: {len(file_paths)}")

# Split files into train and validation sets first (75% train, 25% validation)
train_paths, val_paths, train_labels, val_labels = train_test_split(
    file_paths, labels, test_size=0.25, random_state=42, stratify=labels
)
print(f"✅ Training set: {len(train_paths)} files")
print(f"✅ Validation set: {len(val_paths)} files")

# Encode class labels
encoder = LabelEncoder()
encoder.fit(labels)  # Fit on all labels to ensure consistent encoding
train_labels_encoded = encoder.transform(train_labels)
val_labels_encoded = encoder.transform(val_labels)
print(f"🧠 Classes: {encoder.classes_}")

# Extract features for training set
print("Extracting features for training set...")
X_train = []
for file_path in train_paths:
    try:
        features = extract_features(file_path)
        if features is not None:
            X_train.append(features)
    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
X_train = np.array(X_train)
print(f"✅ Extracted features for {len(X_train)} training files")
print(f"📊 Training data shape: {X_train.shape}")

# Extract features for validation set
print("Extracting features for validation set...")
X_val = []
for file_path in val_paths:
    try:
        features = extract_features(file_path)
        if features is not None:
            X_val.append(features)
    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
X_val = np.array(X_val)
print(f"✅ Extracted features for {len(X_val)} validation files")
print(f"📊 Validation data shape: {X_val.shape}")

# Load and process test data
print("Loading test data...")
X_test, y_test, test_files = load_test_data(test_data_dir, encoder)

if X_test is not None and len(X_test) > 0:
    print(f"✅ Test dataset loaded: {len(X_test)} samples")
    
    # Encode test labels
    y_test_encoded = encoder.transform(y_test)
    print(f"📊 Test data shape: {X_test.shape}")
else:
    print("⚠️ No test data found or could be processed")
    X_test = None
    y_test_encoded = None

# Build 2D CNN model
model = Sequential([
    # First convolutional block
    Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same', 
           input_shape=(X_train.shape[1], X_train.shape[2], X_train.shape[3])),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    
    # Second convolutional block
    Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    
    # Third convolutional block
    Conv2D(128, kernel_size=(3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    
    # Fourth convolutional block
    Conv2D(256, kernel_size=(3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    
    # Flatten and dense layers
    Flatten(),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(len(encoder.classes_), activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary
model.summary()

# Create callbacks for better training
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001)
]

# Train model
history = model.fit(
    X_train, 
    train_labels_encoded,
    epochs=50,
    batch_size=16,
    validation_data=(X_val, val_labels_encoded),
    callbacks=callbacks,
    verbose=1
)

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('raga_classification_history.png')
plt.show()

# Evaluate on test data if available
if X_test is not None and y_test_encoded is not None:
    # Get predictions
    y_pred = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    
    # Calculate metrics
    accuracy = np.sum(y_pred_classes == y_test_encoded) / len(y_test_encoded)
    precision = precision_score(y_test_encoded, y_pred_classes, average='weighted')
    recall = recall_score(y_test_encoded, y_pred_classes, average='weighted')
    f1 = f1_score(y_test_encoded, y_pred_classes, average='weighted')
    
    print(f"🎯 Test Accuracy: {accuracy * 100:.2f}%")
    print(f"🎯 Test Precision: {precision * 100:.2f}%")
    print(f"🎯 Test Recall: {recall * 100:.2f}%")
    print(f"🎯 Test F1 Score: {f1 * 100:.2f}%")
    
    # Create confusion matrix
    cm = confusion_matrix(y_test_encoded, y_pred_classes)
    plt.figure(figsize=(12, 10))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
               xticklabels=encoder.classes_,
               yticklabels=encoder.classes_)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.tight_layout()
    plt.savefig('raga_confusion_matrix.png')
    plt.show()
    
    # Print classification report
    print("\nClassification Report:")
    class_report = classification_report(
        y_test_encoded, y_pred_classes, 
        target_names=encoder.classes_,
        output_dict=True
    )
    
    # Convert to DataFrame for better visualization
    df_report = pd.DataFrame(class_report).transpose()
    print(df_report)
    
    # If test files are available, save predictions to CSV
    if test_files is not None:
        results_df = pd.DataFrame({
            'File': test_files,
            'True Raga': [encoder.classes_[i] for i in y_test_encoded],
            'Predicted Raga': [encoder.classes_[i] for i in y_pred_classes],
            'Correct': y_test_encoded == y_pred_classes
        })
        results_df.to_csv('raga_test_predictions.csv', index=False)
        print("Prediction results saved to 'raga_test_predictions.csv'")

# Save the model
model.save('raga_classification_2d_cnn.h5')
print("Model saved as 'raga_classification_2d_cnn.h5'")
```

```python
import os
import librosa
import warnings
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder, StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report, confusion_matrix, f1_score, precision_score, recall_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Suppress warnings
warnings.filterwarnings("ignore")

# Feature extraction functions for 2D CNN
def extract_melspectrogram(y, sr, n_mels=128, n_fft=2048, hop_length=512):
    """Extract mel spectrogram features suitable for 2D CNN"""
    mel_spec = librosa.feature.melspectrogram(
        y=y, sr=sr, n_mels=n_mels, n_fft=n_fft, hop_length=hop_length
    )
    # Convert to dB scale
    mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)
    return mel_spec_db

def extract_mfcc_2d(y, sr, n_mfcc=20, n_fft=2048, hop_length=512):
    """Extract MFCC features suitable for 2D CNN"""
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=n_mfcc, n_fft=n_fft, hop_length=hop_length)
    # Add delta and delta-delta features
    mfcc_delta = librosa.feature.delta(mfcc)
    mfcc_delta2 = librosa.feature.delta(mfcc, order=2)
    # Stack together to form a 3-channel image
    return np.stack([mfcc, mfcc_delta, mfcc_delta2], axis=-1)

def extract_chroma_2d(y, sr, n_chroma=12, n_fft=2048, hop_length=512):
    """Extract chromagram features suitable for 2D CNN"""
    chroma = librosa.feature.chroma_stft(y=y, sr=sr, n_chroma=n_chroma, n_fft=n_fft, hop_length=hop_length)
    return chroma

# Combine features for 2D CNN
def extract_features(file_path, fixed_length=256):
    """Extract 2D features for CNN processing with fixed time dimension"""
    y, sr = librosa.load(file_path, sr=22050)  # Fixed sample rate for consistency
    
    # Use a fixed duration by either padding or trimming
    target_length = sr * 10  # Target 10 seconds of audio
    if len(y) < target_length:
        # Pad if shorter
        y = np.pad(y, (0, target_length - len(y)), 'constant')
    else:
        # Trim if longer - take a segment from the middle
        start = len(y) // 2 - target_length // 2
        y = y[start:start + target_length]
    
    # Extract mel spectrogram (main feature for 2D CNN)
    mel_spec = extract_melspectrogram(y, sr, n_mels=128)
    
    # Ensure consistent time dimension
    if mel_spec.shape[1] > fixed_length:
        mel_spec = mel_spec[:, :fixed_length]
    elif mel_spec.shape[1] < fixed_length:
        # Pad with zeros if shorter
        padding = fixed_length - mel_spec.shape[1]
        mel_spec = np.pad(mel_spec, ((0, 0), (0, padding)), 'constant')
    
    # Add channel dimension for CNN
    mel_spec = mel_spec.reshape(mel_spec.shape[0], mel_spec.shape[1], 1)
    
    return mel_spec

# Get file paths and labels from dataset directory
def get_file_paths_and_labels(data_dir):
    file_paths = []
    labels = []
    
    for raga in os.listdir(data_dir):
        raga_path = os.path.join(data_dir, raga)
        if not os.path.isdir(raga_path):
            continue
        
        print(f"Finding files for raga: {raga}")
        file_count = 0
        
        for file in os.listdir(raga_path):
            if file.endswith((".wav", ".mp3")) and ".DS_Store" not in file:
                file_path = os.path.join(raga_path, file)
                file_paths.append(file_path)
                labels.append(raga)
                file_count += 1
        
        print(f"  - Found {file_count} files for {raga}")
    
    return file_paths, labels

# Set the data directory
data_dir = "../dataset/"

# Get file paths and labels for all data
file_paths, labels = get_file_paths_and_labels(data_dir)
print(f"✅ Total files found: {len(file_paths)}")

# First split: separate out the test set (20% of total data)
train_val_paths, test_paths, train_val_labels, test_labels = train_test_split(
    file_paths, labels, test_size=0.2, random_state=42, stratify=labels
)

# Second split: divide remaining data into train (75% of train_val = 60% of total) and validation (25% of train_val = 20% of total)
train_paths, val_paths, train_labels, val_labels = train_test_split(
    train_val_paths, train_val_labels, test_size=0.25, random_state=42, stratify=train_val_labels
)

print(f"✅ Training set: {len(train_paths)} files ({len(train_paths)/len(file_paths)*100:.1f}%)")
print(f"✅ Validation set: {len(val_paths)} files ({len(val_paths)/len(file_paths)*100:.1f}%)")
print(f"✅ Test set: {len(test_paths)} files ({len(test_paths)/len(file_paths)*100:.1f}%)")

# Encode class labels
encoder = LabelEncoder()
encoder.fit(labels)  # Fit on all labels to ensure consistent encoding
train_labels_encoded = encoder.transform(train_labels)
val_labels_encoded = encoder.transform(val_labels)
test_labels_encoded = encoder.transform(test_labels)
print(f"🧠 Classes: {encoder.classes_}")

# Extract features for training set
print("Extracting features for training set...")
X_train = []
for file_path in train_paths:
    try:
        features = extract_features(file_path)
        if features is not None:
            X_train.append(features)
    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
X_train = np.array(X_train)
print(f"✅ Extracted features for {len(X_train)} training files")
print(f"📊 Training data shape: {X_train.shape}")

# Extract features for validation set
print("Extracting features for validation set...")
X_val = []
for file_path in val_paths:
    try:
        features = extract_features(file_path)
        if features is not None:
            X_val.append(features)
    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
X_val = np.array(X_val)
print(f"✅ Extracted features for {len(X_val)} validation files")
print(f"📊 Validation data shape: {X_val.shape}")

# Extract features for test set
print("Extracting features for test set...")
X_test = []
test_files = []  # Store filenames for later reference
for file_path in test_paths:
    try:
        features = extract_features(file_path)
        if features is not None:
            X_test.append(features)
            test_files.append(os.path.basename(file_path))
    except Exception as e:
        print(f"⚠️ Error processing {file_path}: {e}")
X_test = np.array(X_test)
print(f"✅ Extracted features for {len(X_test)} test files")
print(f"📊 Test data shape: {X_test.shape}")

# Build 2D CNN model
model = Sequential([
    # First convolutional block
    Conv2D(32, kernel_size=(3, 3), activation='relu', padding='same', 
           input_shape=(X_train.shape[1], X_train.shape[2], X_train.shape[3])),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    
    # Second convolutional block
    Conv2D(64, kernel_size=(3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    
    # Third convolutional block
    Conv2D(128, kernel_size=(3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    
    # Fourth convolutional block
    Conv2D(256, kernel_size=(3, 3), activation='relu', padding='same'),
    BatchNormalization(),
    MaxPooling2D(pool_size=(2, 2)),
    
    # Flatten and dense layers
    Flatten(),
    Dense(256, activation='relu'),
    BatchNormalization(),
    Dropout(0.5),
    Dense(len(encoder.classes_), activation='softmax')
])

# Compile model
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Print model summary
model.summary()

# Create callbacks for better training
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=0.00001)
]

# Train model
history = model.fit(
    X_train, 
    train_labels_encoded,
    epochs=5,
    batch_size=16,
    validation_data=(X_val, val_labels_encoded),
    callbacks=callbacks,
    verbose=1
)

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Validation')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.savefig('raga_classification_history.png')
plt.show()

# Evaluate on test data
# Get predictions
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Calculate metrics
accuracy = np.sum(y_pred_classes == test_labels_encoded) / len(test_labels_encoded)
precision = precision_score(test_labels_encoded, y_pred_classes, average='weighted')
recall = recall_score(test_labels_encoded, y_pred_classes, average='weighted')
f1 = f1_score(test_labels_encoded, y_pred_classes, average='weighted')

print(f"🎯 Test Accuracy: {accuracy * 100:.2f}%")
print(f"🎯 Test Precision: {precision * 100:.2f}%")
print(f"🎯 Test Recall: {recall * 100:.2f}%")
print(f"🎯 Test F1 Score: {f1 * 100:.2f}%")

# Create confusion matrix
cm = confusion_matrix(test_labels_encoded, y_pred_classes)
plt.figure(figsize=(12, 10))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
           xticklabels=encoder.classes_,
           yticklabels=encoder.classes_)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.tight_layout()
plt.savefig('raga_confusion_matrix.png')
plt.show()

# Print classification report
print("\nClassification Report:")
class_report = classification_report(
    test_labels_encoded, y_pred_classes, 
    target_names=encoder.classes_,
    output_dict=True
)

# Convert to DataFrame for better visualization
df_report = pd.DataFrame(class_report).transpose()
print(df_report)

# Save test predictions to CSV
results_df = pd.DataFrame({
    'File': test_files,
    'True Raga': [encoder.classes_[i] for i in test_labels_encoded],
    'Predicted Raga': [encoder.classes_[i] for i in y_pred_classes],
    'Correct': test_labels_encoded == y_pred_classes
})
results_df.to_csv('raga_test_predictions.csv', index=False)
print("Prediction results saved to 'raga_test_predictions.csv'")

# Save the model
model.save('raga_classification_2d_cnn_new.h5')
print("Model saved as 'raga_classification_2d_cnn_new.h5'")
```
